First, we import my models and define them : objective reaction and solver. 
In the publication, the solver used is CPLEX, but in order to make this walkthrough accessible, let's use glpk. 

```python
host = cobra.io.read_sbml_model("resources/enterocyte_ori.xml")
host.objective = host.reactions.get_by_id('biomass_reactionIEC01b')
host.solver = 'cplex'

b = cobra.io.read_sbml_model("resources/embl_gems/models/l/lactobacillus/Lactobacillus_rhamnosus_GG_GG_ATCC_53103.xml")
b.solver = "cplex"
bacteria_id = 'Lactobacillus rhamnosus GG'
```

Then, we build the metabolic_dict. The keys are tuples (metabolite_id, model) and the value is the metabolite_id that will be used in the ecosystem model. It enables to join metablites named differently accross models into a common id in the ecosystem model.
In this case, the namespaces are the same. The metabolic_dict is therefore a simple enumeration of all metabolites and their id. 

```python
enteromet = []
for met in host.metabolites:
    enteromet.append((met.id, host))
bacterimet = []
for met in b.metabolites:
    bacterimet.append((met.id, b))

all_metabolites = list(set(enteromet + bacterimet))
metabolic_dict = {x:x[0] for x in all_metabolites}
```

## Diet constraint
The nutritionnal constraint comes with the redefinition of the bounds of the exchange reactions for chosen metabolites. The lower bound represents the influx of the metabolite in the pool compartment.

First, we have to constrain the exchanges for the metabolites with the blood :

```python
medium_blood = pd.read_csv("resources/AAD_notebook.tsv", sep="\t", index_col = 0)

for reaction in medium_blood.index:
    host.reactions.get_by_id(reaction).bounds = (medium_blood.loc[reaction,'lb'],medium_blood.loc[reaction,'ub'])

#diet.medium_2(model = host, medium_lumen = diet_param, host_choice = host_choice, block = False, mocba=True, semiconstrained = False)
```

Then, we will constrain the exchanges with the lumen. We need to define this constraint by creating a dictionnary with 
 - as keys : the id of the pool metabolites
 - as values : a tuple (lb, ub)
 
If a metabolite is part of the ecosystem pool, but not in this dictionnary, the bounds will be (-1000, 1000)

We begin with constraining both models with the Western diet, and optimize their biomass separately :

```python
with open("/home/e158401a/Documents/pareto_analysis/WD_host_LGG.pickle", "rb") as fp:   # Unpickling
    diet_dict = pickle.load(fp)

#with open("resources/diet_dict_LGG_Western_diet.pickle", "rb") as fp:   # Unpickling
#    diet_dict = pickle.load(fp)
       
for ex in host.exchanges:
    met_e = list(ex.metabolites.keys())[0].id
    if met_e in list(diet_dict.keys()):
        ex.lower_bound = diet_dict[met_e][0]

host_biomass_alone = host.optimize().objective_value
print('host max biomass when alone in a WD : ',host_biomass_alone)

for ex in b.exchanges:
    met_e = list(ex.metabolites.keys())[0].id
    if met_e in list(diet_dict.keys()):
        ex.lower_bound = diet_dict[met_e][0]
        
LGG_biomass_alone = b.optimize().objective_value
print('LGG max biomass when alone in a WD : ',LGG_biomass_alone)
```

Now, we have to prepare te models to be joined in an ecosystem. We need to unconstrain the original exchange reactions, which will become transport reactions between the organism's external compartment and the ecosystem's pool compartment.

```python
for ex in host.exchanges:
    ex.bounds = (-1000,1000)
    
for ex in b.exchanges:
    ex.bounds = (-1000,1000)
```

We then join the two models into an ecosystem whith the metabolites named based on the metabolic_dict and the pool exchange reactions constrained based on the diet_dict. 

```python
from mocbapy.EcosystemModel import create_model, bensolve_default_options
ecosys = create_model(model_array=[host, b], metabolic_dict=metabolic_dict, diet = diet_dict)
```

Now that the ecosystem is created, we will explore the pareto. To do that, we use mocbapy to do a mofba (multi-objective flux balance analysis)

```python
import mocbapy.analysis

bensolve_opts = bensolve_default_options()
bensolve_opts['message_level'] = 0
sol_mofba = mocbapy.analysis.mo_fba(ecosys, options=bensolve_opts)
```

sol_mofba contains the extreme points of the pareto front. We normalize these points with the growth of each organism alone, and verify the conditions to infer interaction categories. Here, The ["0", "0", "0"] is the initialization, where 0 is "False" basd on the category definition in methods.

```python
x = []
y = []
maxi_host = (host_biomass_alone, 0)
maxi_b = (0, LGG_biomass_alone)
category = ["0","0","0"]
for i in range(len(sol_mofba.Primal.vertex_value)):
    if sol_mofba.Primal.vertex_type[i] == 1: #1 means point when 0 is a vector
        x.append(sol_mofba.Primal.vertex_value[i][0]/host_biomass_alone)
        y.append(sol_mofba.Primal.vertex_value[i][1]/LGG_biomass_alone) 
        if sol_mofba.Primal.vertex_value[i][0] > maxi_host[0]: #Is the growth of the enterocyte at this point on the pareto higher than the previously saved maximal growth ?
            maxi_host = sol_mofba.Primal.vertex_value[i] #Then it's our new max
            category[0] = "1"
        if sol_mofba.Primal.vertex_value[i][1] > maxi_b[1]:
            maxi_b = sol_mofba.Primal.vertex_value[i]
            category[1] = "1"
if tuple(maxi_host) == tuple(maxi_b) and (category[0] == "1" and category[1] == "1"):
    category[2] = "1"
category=''.join(category)
```

The pareto keeps only the optimal solutions for the ecosystem. The growths of the organisms alone are not optimal, but they are needed to see the impact of the ecosystem on the growth of the organisms. We add them to the pareto points (singular in this case).

```python
xy = pd.DataFrame({'x': x, 'y': y})
xy = xy.append({'x' : 1.00001, 'y' : -0.00001}, ignore_index=True)
xy.loc[-1] = [-0.00001, 1.00001] #To put it first in the df
xy.index = xy.index + 1  # shifting index
xy = xy.sort_index()
xy
```

Now, we calculate the area under the curve (AUC) of this normalized pareto front

```python
#To measure area under (AUC) the curve when the x is neither monotonic increasing nor monotonic decreasing
from sklearn import metrics

AUC1 = metrics.auc(x = xy['x'][0:-1], y = xy['y'][0:-1])
AUC2 = metrics.auc(x = xy['x'][-2:], y = xy['y'][-2:])
S = AUC1 - AUC2 - 1
```

The sign of the AUC is my condition 1 to determine the category :

```python
if S < -0.0001 and category == "000":
    category = "-000"
elif S > -0.0001 and S < 0.0001: #I allow a margin for neutrality
    category = "+000"   
```

Finally, here is the appearance of the pareto front

```python
import matplotlib.pyplot as plt

plt.title("Pareto front of host - bacteria metabolic interaction\n")
plt.xlabel("enterocyte's maintenance")
plt.ylabel(bacteria_id+"'s growth")
plt.plot(xy['x'], xy['y'], '#ff0000', linestyle="-")
plt.fill_between(xy['x'], xy['y'], color = "#f08c8c30")
plt.axhline(y = 1, color = '#1155cc', linestyle = '--', linewidth = 1)
plt.axvline(x = 1, color = '#1155cc', linestyle = '--', linewidth = 1)
plt.show()
```

# Sampling
First, to be able to use the Flux Balance Analysis (FBA) I transform my ecosystem model into a format compatible with cobra

```python
model = cobra.Model('ecosys')
for m in range(len(ecosys.sysmetabolites)):
    model.add_metabolites(cobra.Metabolite(ecosys.sysmetabolites[m]))
for r in range(len(ecosys.sysreactions)):
    reaction = cobra.Reaction(ecosys.sysreactions[r])
    reaction.lower_bound = ecosys.lb[r]
    reaction.upper_bound = ecosys.ub[r]
    dict_metabolites = {}
    model.add_reactions([reaction])
    for m in range(len(ecosys.sysmetabolites)):
        if ecosys.Ssigma[m,r] != 0:
            dict_metabolites[ecosys.sysmetabolites[m]] = ecosys.Ssigma[m,r]
    reaction.add_metabolites(dict_metabolites)
```

```python
fba = model.optimize()
sampling_dict = {}
sampling_dict["reactions"] = fba.fluxes.index

nbpts_sampling = 10000 #Number of sample to take. 100000 in the study, but only 100 here to be quicker. 
```

```python
import math

sumdist = 0        
for p in xy.index[1:]: #I divide my pareto into segments between 2 points. I denormalize it so that my pareto point can become objective constraints
    x1 = xy.loc[p-1, 'x'] * host_biomass_alone
    y1 = xy.loc[p-1, 'y'] * LGG_biomass_alone
    x2 = xy.loc[p, 'x'] * host_biomass_alone
    y2 = xy.loc[p, 'y'] * LGG_biomass_alone
    dist = math.hypot(x2-x1, y2-y1)
    sumdist = sumdist + dist
d1pt = nbpts_sampling/sumdist
```

```python
for p in xy.index[1:]:
    print(p)
    x1 = xy.loc[p-1, 'x'] * host_biomass_alone
    y1 = xy.loc[p-1, 'y'] * LGG_biomass_alone
    x2 = xy.loc[p, 'x'] * host_biomass_alone
    y2 = xy.loc[p, 'y'] * LGG_biomass_alone
    dist = math.hypot(x2-x1, y2-y1)
    nbpnts_xy = round(dist*d1pt)
    print(dist)
    print(nbpnts_xy)
    if nbpnts_xy>0:
        dx = (x2-x1)/nbpnts_xy
        dy = (y2-y1)/nbpnts_xy
        for i in range (0, nbpnts_xy):
            pix = x1+dx*i
            piy = y1+dy*i
            model.reactions.get_by_id('biomass_reactionIEC01b:IEC1907').bounds = (pix, pix)
            model.reactions.get_by_id('Growth:'+b.id).bounds = (piy, piy)
            fba = model.optimize() #One sample
            sampling_dict[str(pix)+"_"+str(piy)] = list(fba.fluxes)    
sampling = pd.DataFrame(data=sampling_dict)
sampling.index = sampling["reactions"]
sampling = sampling.T
sampling = sampling.drop("reactions")
```

Keeping only exchange reactions from sampling to lighten the data

```python
exchanges = []
for reac in sampling.columns:
    if reac[0:3] == "EX_":
        exchanges.append(reac)
exchanges.append("biomass_reactionIEC01b:IEC1907")
exchanges.append("Growth:Lactobacillus_rhamnosus_GG_GG_ATCC_53103_xml")
sampling_exchanges = sampling[exchanges]
del sampling
```

# Correlation

```python
sampling_exchanges = sampling_exchanges.astype(float)
correlation_reactions = sampling_exchanges.corr(method ='spearman')
correlation_reactions = correlation_reactions.fillna(0)
```

# Exchanged metabolites

```python
e_suffixe = "_e:IEC1907"
b_suffixe = "_e:Lactobacillus_rhamnosus_GG_GG_ATCC_53103_xml"

def short_name(old_name):
    new_name = old_name.replace(":IEC1907", ":e")
    new_name = new_name.replace(b_suffixe, ":b")
    return new_name

def only_met(reac):
    met = reac.replace("_e:IEC1907", "")
    met = met.replace("_e:"+b_suffixe[3:], "")
    met = met.replace("EX_", "")
    return met

def oppositeSigns(x, y): #returns boolean. 1 if x y opposite sign, 0 otherwhise
    if x < 0:
        return (y > 0) # no <= Because I want a net import 
    elif x > 0:
        return (y < 0)
    else:
        return False
```

```python
bact_exchanges = []
for ex in b.exchanges : bact_exchanges.append(ex.id+b_suffixe[2:])

potential_exchange = {}
done = []
for reac in bact_exchanges:
    met = only_met(reac)
    exmet_e = "EX_"+met+e_suffixe
    exmet_b = "EX_"+met+b_suffixe
    if exmet_e in correlation_reactions.index and exmet_b in correlation_reactions.index:
        #if not math.isnan(correlation_reaction.loc[exmet_e,exmet_b]) and sum(sampling[exmet_e])!=0 and sum(sampling[exmet_b])!=0 and abs(correlation_reaction.loc[exmet_e,exmet_b]) >= 0.5:
        if sum(sampling_exchanges[exmet_e])!=0 and sum(sampling_exchanges[exmet_b])!=0 and correlation_reactions.loc[exmet_e,exmet_b] <= -0.5:
            exchange = 0 
            way = ["0","0"]
            for s in sampling_exchanges.index:
                if oppositeSigns(sampling_exchanges.loc[s, exmet_e], sampling_exchanges.loc[s, exmet_b]) :
                    exchange = exchange+1
                    if sampling_exchanges.loc[s, exmet_e] <0 :
                        way[0] = "1"
                    elif sampling_exchanges.loc[s, exmet_b]<0:
                        way[1] = "1"
            if exchange >1 and short_name(exmet_e)+"|||"+short_name(exmet_b) not in done:
                potential_exchange[short_name(exmet_e)] = [exchange, way]
                A = sampling_exchanges[exmet_e]
                B = sampling_exchanges[exmet_b] 
                plt.plot(A, "#e06666", label='enterocyte')
                plt.plot(B, "#93c47d", label='bacteria')
                if met+"_e" in list(diet_dict.keys()):
                    plt.axhline(y=diet_dict[met+"_e"][0], color='k', linestyle='--')
                else:
                    plt.axhline(y=0, color='k', linestyle='--')
                plt.tick_params(
                    axis='x',
                    which='both',    
                    bottom=False,     
                    top=False,         
                    labelbottom=False) 
                plt.title("evolution of "+met+" exchanges on the pareto front")
                plt.legend()
                plt.show()
            done.append(short_name(exmet_e)+"|||"+short_name(exmet_b))
```
